```python
import geopandas as gpd
import rasterio
import rasterio.mask
import pandas as pd
import numpy as np
import re
import os

SHAPEFILE_PATH = 'UCL_2021_AUST_GDA2020_SHP/UCL_2021_AUST_GDA2020.shp'
TIFF_BASE_PATH = 'Australia_GISdata_LTAy_AvgDailyTotals/monthly/PVOUT_{month}.tif'
OUTPUT_CSV = 'australia_solar_output_final.csv'

my_cities = [ 'Adelaide', 'Albany', 'Albury', 'AliceSprings', 'BadgerysCreek', 'Ballarat', 
    'Bendigo', 'Brisbane', 'Cairns', 'Canberra', 'Cobar', 'CoffsHarbour', 
    'Dartmoor', 'Darwin', 'GoldCoast', 'Hobart', 'Katherine', 'Launceston', 
    'Melbourne', 'MelbourneAirport', 'Mildura', 'Moree', 'MountGambier', 
    'MountGinini', 'Newcastle', 'Nhil', 'NorahHead', 'NorfolkIsland', 
    'Nuriootpa', 'PearceRAAF', 'Penrith', 'Perth', 'PerthAirport', 'Portland', 
    'Richmond', 'Sale', 'SalmonGums', 'Sydney', 'SydneyAirport', 'Townsville', 
    'Tuggeranong', 'Uluru', 'WaggaWagga', 'Walpole', 'Watsonia', 'Williamtown', 
    'Witchcliffe', 'Wollongong', 'Woomera']

# Manual Mapping to bridge the gap between Weather Names and ABS Names
MANUAL_MAPPING = {"BadgerysCreek": "Sydney",
    "MelbourneAirport": "Melbourne",
    "SydneyAirport": "Sydney",
    "PerthAirport": "Perth",
    "Penrith": "Sydney",
    "Watsonia": "Melbourne",
    "Tuggeranong": "Canberra",
    "MountGinini": "Canberra",
    "PearceRAAF": "Bullsbrook",
    "Williamtown": "Newcastle",
    "Nhil": "Nhill",  
    "NorahHead": "Central Coast"}


def get_search_name(name):
    """Translates weather station names to searchable ABS names"""
    if name in MANUAL_MAPPING:
        return MANUAL_MAPPING[name]
    return re.sub(r"(\w)([A-Z])", r"\1 \2", name).strip() # Adds space before capital letters (e.g., AliceSprings -> Alice Springs)

def calculate_area_average(geom, month_idx):
    """Performs the raster masking and returns the mean solar value"""
    month_str = str(month_idx).zfill(2)
    tif_path = TIFF_BASE_PATH.format(month=month_str)
    
    if not os.path.exists(tif_path):
        return None

    with rasterio.open(tif_path) as src:
        try:
            out_image, out_transform = rasterio.mask.mask(src, [geom], crop=True) #Mask the raster using the city polygon
            data = out_image[0]
            valid_pixels = data[data > 0] # ilter out invalid pixels (0 or negative)
            
            if valid_pixels.size == 0:
                return None
                
            return np.mean(valid_pixels)
        except Exception:
            return None


def main():
    gdf = gpd.read_file(SHAPEFILE_PATH)
    if gdf.crs != "EPSG:4326":
        gdf = gdf.to_crs("EPSG:4326")

    results = []
    for city in my_cities:
        search_term = get_search_name(city)
        # Look for the city in the UCL_NAME21 column:
        match = gdf[gdf['UCL_NAME21'].str.contains(search_term, case=False, na=False)]
        # If no match:        
        if match.empty:
            print(f"  [!] Missing: {city} (searched as '{search_term}')")
            continue
            
        city_poly = match.geometry.iloc[0]
        abs_name = match['UCL_NAME21'].iloc[0]

        for m in range(1, 13):
            avg_pvout = calculate_area_average(city_poly, m)
            
            if avg_pvout is not None:
                # Calculation: Average Daily PVOUT * Typical Efficiency (0.2 kW/m2)
                output_per_m2 = avg_pvout * 0.2
                
                results.append({"City_Input": city, "ABS_Area_Name": abs_name, "Month": m, "Daily_PVOUT_kWh_kWp": round(avg_pvout, 4),
                    "Solar_Output_kWh_m2": round(output_per_m2, 4)})
        print(f"  [+] Completed: {city}")

    final_df = pd.DataFrame(results)
    final_df.to_csv(OUTPUT_CSV, index=False)
    return final_df

if __name__ == "__main__":
    main()
```
